```python
import pandas as pd
import numpy as np
```

```python
weather_total = pd.read_csv('final_weather_table.txt',low_memory=False)
weather_total.head()
```

```python
#WindSpeed
windspeed = weather_total.iloc[:,[1,2,3,-1]]
windspeed.replace('M',np.nan,inplace=True)
windspeed.dropna(inplace=True)
windspeed.head(10)
```

```python
#revise the type of columns
windspeed['YYMM'] = windspeed['YYMM'].astype(str)
windspeed['Avgwindspeed'] = windspeed['Avgwindspeed'].astype(float)
```

```python
#for loop for all 51 states
statelist = ['AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DC', 'DE', 'FL', 'GA', 'HI', 
             'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME', 'MI', 'MN', 
             'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM', 'NV', 'NY', 'OH', 
             'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX', 'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
state_total = pd.DataFrame()

#Run for all 51 states
for item in statelist:
    name = item #remain the name of state as str
    item = windspeed[windspeed['State'].isin([item])] #extract dataframe of a specific state
    
    #for all the year in a specific state

    item.reset_index(inplace=True)
    item = item.iloc[:,1:]

    year_list = ['1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006',
                '2007','2008','2009','2010','2011','2012','2013','2014','2015','2016']
    avg_list = []
    avg_windspeed = pd.DataFrame()
    for i in year_list:
        year = item[item['YYMM'].str.contains(i)]#find all the data in a specific year
        avg = sum(list(year['Avgwindspeed']))/len(list(year['Avgwindspeed']))
        avg_list.append(avg)

    #Create a list of state containing 21 items of AZ
    state = pd.Series(np.zeros(21))
    state.replace('0.0',name,inplace=True)

    avg_windspeed['State'] = state
    avg_windspeed['Year'] = year_list
    avg_windspeed['Avgwindspeed'] = avg_list
    
    state_total = state_total.append(avg_windspeed,ignore_index=True)
```

```python
state_total.to_csv('state_total_windspeed.csv')
```

```python
#Precip
precip = weather_total.iloc[:,[1,2,3,-2]]
precip.replace('M',np.nan,inplace=True)
precip.replace('T','0.0',inplace=True)
precip.dropna(inplace=True)
precip.head(20)
```

```python
#revise the type of columns
precip['YYMM'] = precip['YYMM'].astype(str)
precip['TotalMonthlyPrecip'] = precip['TotalMonthlyPrecip'].astype(float)
```

```python
state_total_precip.to_csv('state_total_precip.csv')
```
