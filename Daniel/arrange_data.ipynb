```python
import pandas as pd
import itertools as it
import numpy as np
```

```python
data = pd.read_csv('../Original_Data/annual_generation_source_state.csv')
```

```python
data = data.drop(['Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9'],axis = 1)
```

```python
data = data[data.STATE != '  ']
data = data[data.STATE != 'US-TOTAL']
data = data[data.STATE != 'US-Total']
```

```python
data
```

```python
yearlist = []
for key, group in it.groupby(sorted(data.YEAR)):
    yearlist.append(key)
print(yearlist)
```

```python
statelist = []
for key, group in it.groupby(sorted(data.STATE)):
    statelist.append(key)
print(statelist)
```

```python
esourcelist = []
for key, group in it.groupby(sorted(data['ENERGY SOURCE'])):
    esourcelist.append(key)
print(esourcelist)
```

```python
arrangedata = pd.DataFrame(index = np.arange(len(yearlist)*len(statelist)),columns = (['Year','State'] + esourcelist))
year = []
for i in yearlist:
    year += list(np.full(len(statelist),i,dtype = int))
arrangedata.Year = year
state = []
for i in range(len(yearlist)):
    state += statelist
arrangedata.State = state
```

```python
arrangedata
```

```python
for m in range(len(yearlist)*len(statelist)):
    for i in yearlist:
        for j in statelist:
            for k in esourcelist:
                esourcesum = 0
                for l in range(data.shape[0]):
                    if data.loc[l].YEAR == i and data.loc[l].STATE == j and data.loc[l]['ENERGY SOURCE'] == k:
                        esourcesum += int(data.loc[l]['GENERATION (Megawatthours)'].replace(',',''))
                        arrangedata.loc[m]['Year'] = i
                        arrangedata.loc[m]['State'] = j
                        arrangedata.loc[m][k] = esourcesum
```
