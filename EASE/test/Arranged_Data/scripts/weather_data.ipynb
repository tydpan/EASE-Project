```python
import pandas as pd
import numpy as np
import math
import os
import glob
```

```python
#station data
weather_station = pd.read_csv("../Original_Data/Final_weather_data/Station_list/station_list.txt",delimiter='|')
weather_table = weather_station[['WBAN','State','Latitude','Longitude']]
```

```python
#try out dataset from 199607
#building block for dataframe processing
dailyavg9607 = pd.read_csv("../Original_Data/Final_weather_data/200706dailyavg.txt",delimiter=',')
readindata = dailyavg9607.rename(columns = {'Wban Number':'WBAN'}).iloc[:,[0,6,26,-3]]
WBAN_copy = readindata['WBAN']
weather_date = pd.DataFrame(dailyavg9607.iloc[:,1]/float(100))
temp_list = []
for i in weather_date.iloc[:,0]:
    temp_list.append(math.floor(i))
    
afprocessedwd = pd.DataFrame(temp_list)
result_data = pd.concat([readindata, afprocessedwd], axis=1)
result_data.columns = ['WBAN','AvgTemp','TotalMonthlyPrecip','Avgwindspeed','YYMM']
result_data = result_data[['YYMM','WBAN','AvgTemp','TotalMonthlyPrecip','Avgwindspeed']]
final_weather_table = pd.merge(left=weather_table,right=result_data, how='left', left_on='WBAN', right_on='WBAN')
final_weather_table.sort_values('State', inplace=True, ascending=True)
final_weather_table.reset_index(inplace=True, drop=True)
final_weather_table
```

```python
#adding iteration over all the files in the folder
path =r'../Original_Data/Final_weather_data'
allFiles = glob.glob(path + "/*.txt")
weather_frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    read_in = pd.read_csv(file_,index_col=None, header=0)
    readindata = read_in.rename(columns = {'Wban Number':'WBAN'}).iloc[:,[0,6,26,-3]]
    weather_date = pd.DataFrame(read_in.iloc[:,1]/float(100))
    temp_list = []
    for i in weather_date.iloc[:,0]:
        temp_list.append(math.floor(i))
    # concat the dataframes and give the columns' name
    # sort the dataframe by state and reset index
    afprocessedwd = pd.DataFrame(temp_list)
    result_data = pd.concat([readindata, afprocessedwd], axis=1)
    result_data.columns = ['WBAN','AvgTemp','TotalMonthlyPrecip','Avgwindspeed','YYMM']
    result_data = result_data[['YYMM','WBAN','AvgTemp','TotalMonthlyPrecip','Avgwindspeed']]
    final_weather_table = pd.merge(left=result_data,right=weather_table, how='left', left_on='WBAN', right_on='WBAN')
    final_weather_table.sort_values('State', inplace=True, ascending=True)
    final_weather_table.reset_index(inplace=True, drop=True)
    final_weather_table = final_weather_table[['YYMM','State','WBAN','Latitude','Longitude','AvgTemp','TotalMonthlyPrecip','Avgwindspeed']]
    list_.append(final_weather_table)
weather_frame = pd.concat(list_)
weather_frame.head(20)
```

```python
path =r'../Original_Data/Final_weather_data/More'
allFiles07to16 = glob.glob(path + "/*.txt")
weather_frame07to16 = pd.DataFrame()
list07to16_ = []
for file_ in allFiles07to16:
    read_in = pd.read_csv(file_, index_col=None, header=0, error_bad_lines=False)
    result_data = read_in.rename(columns = {'Wban Number':'WBAN'}).iloc[:,[0,1,6,26,-3]]
    final_weather_table = pd.merge(left=weather_table,right=result_data, how='left', left_on='WBAN', right_on='WBAN')
    final_weather_table.columns = ['WBAN','State','Latitude','Longitude','YYMM','AvgTemp','TotalMonthlyPrecip','Avgwindspeed']
    final_weather_table.sort_values('State', inplace=True, ascending=True)
    final_weather_table.reset_index(inplace=True, drop=True)
    final_weather_table = final_weather_table[['YYMM','State','WBAN','Latitude','Longitude','AvgTemp','TotalMonthlyPrecip','Avgwindspeed']]
    list07to16_.append(final_weather_table)
weather_frame07to16 = pd.concat(list07to16_)
weather_frame07to16.head(20)
```

```python
# concat two dataframes and output it as a whole
df = [weather_frame, weather_frame07to16]
output_weather_data = pd.concat(df)
output_weather_data
```

```python
# drop all nan terms and reset index
output_weather_data = output_weather_data.dropna(subset=['YYMM','State'])
output_weather_data = output_weather_data.set_index(np.arange(output_weather_data.shape[0]))
```

```python
# convert the type of all elements in list to string
def delwhitespace(a_list):
    for i in range(len(a_list)):
        a_list[i] = str(a_list[i]).strip()
    return a_list

# pass the missing data, small data, and nan terms, convert the type of useful value to float
def chg_to_float(a_list):
    for i in range(len(a_list)):
        if a_list[i] == 'M' or a_list[i] == 'T' or a_list[i] == 'NaN' or a_list[i] == '':
            pass
        else:
            a_list[i] = float(a_list[i])
    return a_list

output_weather_data.TotalMonthlyPrecip = output_weather_data.TotalMonthlyPrecip.replace('*****','M')

for i in list(output_weather_data.columns)[5:]:
    output_weather_data[i] = chg_to_float(delwhitespace(list(output_weather_data[i])))
    
output_weather_data.YYMM = output_weather_data.YYMM.astype(int)
output_weather_data
```

```python
output_weather_data.to_csv('final_weather_table.txt')
```
