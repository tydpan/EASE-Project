```python
import pandas as pd
import itertools as it
import numpy as np
```

```python
# read data and drop unnamed column
data = pd.read_csv('arranged_esource')
data = data.drop(['Unnamed: 0'],axis = 1)
# combine resources and drop the unwanted ones, then rename the columns
data.Other = data.Geothermal + data['Other Biomass'] + data['Other Gases'] + data['Pumped Storage'] + data['Wood and Wood Derived Fuels']
data = data.drop(['Geothermal','Other Biomass','Other Gases','Pumped Storage','Wood and Wood Derived Fuels'],axis = 1)
data = data[['Year','State','Coal','Natural Gas','Petroleum','Nuclear','Hydroelectric Conventional','Solar Thermal and Photovoltaic','Wind','Other','Total']]
data = data.rename(columns = {'Petroleum':'Petro','Natural Gas':'NG','Hydroelectric Conventional':'Hydro','Solar Thermal and Photovoltaic':'Solar'})
```

```python
# extract specific columns 
data_percent = pd.DataFrame()
data_percent[data.columns[0:2]] = data.iloc[:,0:2]
data_percent[data.columns[2:-1]] = data.iloc[:,2:-1].div(data.Total,axis = 0)
```

```python
data_percent.to_csv('esource_percent.csv')
```
